# 线图

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()  # 累加
ts.plot()   # Series直接调用plot
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index,
                  columns=list('ABCD'))
df = df.cumsum()
plt.figure()
df.plot()     # 绘制所有列
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3
df3.plot(x='A', y='B')  # 指定xy为某两列
```

histogram 频次统计    

## DataFrame.plot.hist() 和 Series.plot.hist()   

```python
df4 = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 1,
    },
    columns=["a", "b", "c"],
)
plt.figure()
df4.plot.hist(alpha=0.5)
```

```python
df4['a'].plot.hist(orientation="horizontal", bins=20, cumulative=True)
```

## DataFrame.hist()
常用颜色简写（ color = 'b' ）  
b: 蓝色（blue）

g: 绿色（green）

r: 红色（red）

c: 蓝绿色（cyan）

m: 紫红色（magenta）

y: 黄色（yellow）

k: 黑色（black）

w: 白色（white）

```python
# dataframe的每一列一个子图
plt.figure();
df4.diff().hist(color="y", alpha=0.5, bins=50);  # bins 分几段
```

## Series.hist()  

```python
plt.figure();
df4["a"].diff().hist();
```

```python
# 使用by参数 分组统计
data = pd.Series(np.random.randn(1000))
data.hist(by=np.random.randint(0, 4, 1000), figsize=(6, 4));  
```

```python
data = pd.DataFrame(
    {
        "a": np.random.choice(["x", "y", "z"], 1000),
        "b": np.random.choice(["e", "f", "g"], 1000),
        "c": np.random.randn(1000),
        "d": np.random.randn(1000) - 1,
    },
)
data['c'].hist(by=data['a'].values, figsize=(10, 5))
```

# 箱线图 box

```python
df = pd.DataFrame(np.random.rand(10, 5))
plt.figure();
color = {
    "boxes": "DarkGreen",
    "whiskers": "DarkOrange",
    "medians": "DarkBlue",
    "caps": "Gray",
}
df.plot.box(color=color, sym="r+");
```

四分位数：它是一组数据排序后处于25%和75%位置上的值。
