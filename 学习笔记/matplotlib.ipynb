# axvspan()
绘制一块矩形区域

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls="-.", lw=2, c="c", label="plot figure")

plt.legend()

plt.axhspan(ymin=0.0, ymax=0.5, facecolor="y", alpha=0.3)
plt.axvspan(xmin=4.0, xmax=6.0, facecolor="y", alpha=0.3)
# plt.axvspan(xmin=4.0, xmax=6.0,ymin=0.5, ymax=1, facecolor="y", alpha=0.3)

plt.show()
````

# subplot 和 subplots
subplot直接按位置开始画  
subplots先把画版画好，一次生成多个，用的时候直接调用即可

````python
from matplotlib import pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.1)
s = np.sin(t * np.pi)
plt.subplot(2, 2, 1)  #要生成两行两列，这是第一个图plt.subplot('行','列','编号')
plt.plot(t, s, 'b--')
plt.ylabel('y1')
plt.subplot(2, 2, 2)  #两行两列,这是第二个图
plt.plot(2 * t, s, 'r--')
plt.ylabel('y2')
plt.subplot(2, 2, 3)  #两行两列,这是第三个图
plt.plot(3 * t, s, 'm--')
plt.subplot(2, 2, 4)  #两行两列,这是第四个图
plt.plot(4 * t, s, 'k--')
plt.show()
````

````python
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
figure,ax=plt.subplots(2,2)
ax[0][0].plot(t,s,'r*')
ax[0][1].plot(t*2,s,'b--')
plt.show()
````
