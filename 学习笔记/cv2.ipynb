# 读图显示
cv2.waitKey(0) 等待用户键入

```python
import cv2

img = cv2.imread('../img/lWUlB.jpg')
cv2.imshow('origin', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 直接在jupyter notebook中显示

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
img2 = img[:, :, ::-1]  # 必须为 ::-1
plt.imshow(img2)
```

## 边框处理 cv2.copyMakeBorder

```python
constant = cv2.copyMakeBorder(img2,40,40,40,40,cv2.BORDER_CONSTANT,value=(128,128,128))
plt.imshow(constant)
```

```python
constant = cv2.copyMakeBorder(img2,40,40,40,40,cv2.BORDER_WRAP)
plt.imshow(constant)
```

## 翻转 cv2.flip()
翻转，0 沿X轴翻转（垂直翻转）， 1 沿Y轴翻转（水平翻转）， 负数 两边都翻转

```python
flip = cv2.flip(img2,1)
plt.imshow(flip)
```

## resize()
interpolation 选项所用的插值方法:   

| 参数| 方法描述 |  
| :-------------- | :---------------------------- |  
| INTER_NEAREST  | 最近邻插值                   |  
| INTER_LINEAR   | 双线性插值（默认设置）       |  
| INTER_AREA     | 使用像素区域关系进行重采样。 |  
| INTER_CUBIC    | 4x4像素邻域的双三次插值      |  
| INTER_LANCZOS4 | 8x8像素邻域的Lanczos插值     |  

```python
img_resize = cv2.resize(img2, (300, 300))
print(img_resize.shape)
plt.imshow(img_resize)
```

```python
img_resize = cv2.resize(img2, (0, 0),fx=0.25, fy=0.25, interpolation=cv2.INTER_NEAREST)
print(img_resize.shape)
plt.imshow(img_resize)
```
